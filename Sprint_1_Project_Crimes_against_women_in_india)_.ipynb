### ***Chart 1***

### **Visualizing Yearly Crime Trends by Category**

```python
# Group data by 'Year' and sum up the total number of crimes for each crime category per year

# Define the crime categories of interest to be analyzed
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                 'Assault against Women', 'Assault against Modesty of Women',
                 'Domestic Violence', 'Women Trafficking']

# Group the DataFrame by 'Year' and sum the values of the specified crime categories for each year
yearly_data = crimes_df.groupby('Year')[crime_columns].sum()

# Create a figure for the line plot, with a specified size of 10 by 6 inches
plt.figure(figsize=(10, 6))

# Loop through each crime category and create a line plot for each one
for crime in crime_columns:
    # Use Seaborn's lineplot function to plot crime trends over the years
    sns.lineplot(data=yearly_data, x=yearly_data.index, y=crime, label=crime)

# Set the title of the plot with a font size of 14
plt.title('Yearly Trend of Different Crimes', fontsize=14)

# Set the label for the x-axis as 'Year' with a font size of 12
plt.xlabel('Year', fontsize=12)

# Set the label for the y-axis as 'Number of Crimes' with a font size of 12
plt.ylabel('Number of Crimes', fontsize=12)

# Add a legend to the plot that shows which line corresponds to which crime type
plt.legend(title='Crime Type')

# Display the line plot on the screen
plt.show()
```

**Insight**

There was a dramatic rise in crimes like domestic violence, kidnapping, and assault against women after 2011, suggesting increased reporting, awareness, or actual rises in these crimes. In contrast, dowry deaths and women trafficking have shown relatively stable trends, indicating different crime patterns.


### ***Chart 2***

### **Visualization of States with the Highest Crime Incidents**

```python
# Import necessary libraries for creating interactive widgets and dropdowns
from ipywidgets import interact, Dropdown

# Define the crime categories to analyze
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                 'Assault against Women', 'Assault against Modesty of Women',
                 'Domestic Violence', 'Women Trafficking']

# Calculate the total number of crimes for each record by summing specified crime categories
crimes_df['Total Crimes'] = crimes_df[crime_columns].sum(axis=1)

# Group the data by 'State' and calculate the total crimes per state, sorting in descending order
state_crime_totals = crimes_df.groupby('State')['Total Crimes'].sum().sort_values(ascending=False)

# Get the top 10 states based on total crimes and filter the main DataFrame for these states
top_10_states = state_crime_totals.head(10).index
top_10_data = crimes_df[crimes_df['State'].isin(top_10_states)]

# Define a function to update the bar plot based on the selected crime type from the dropdown
def update_barplot(selected_crime):
    # Group the filtered data by 'State' and sum the occurrences of the selected crime type for the top 10 states
    top_10_crimes = top_10_data.groupby('State')[selected_crime].sum().sort_values(ascending=False)

    # Create a bar plot for the selected crime type
    plt.figure(figsize=(12, 6))
    top_10_crimes.plot(kind='bar', color='purple')
    plt.title(f'Top 10 States by {selected_crime}', fontsize=14)  # Set the title of the plot
    plt.xlabel('State', fontsize=12)  # Label for the x-axis
    plt.ylabel(f'Total {selected_crime} Cases', fontsize=12)  # Label for the y-axis
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.show()  # Display the bar plot

# Create a dropdown widget for selecting crime types, with 'Rape' set as the default value
crime_dropdown = Dropdown(
    options=crime_columns,  # List of options for the dropdown
    value='Rape',           # Default value selected in the dropdown
    description='Crime Type:',  # Label for the dropdown
)

# Create an interactive plot with the dropdown
interact(update_barplot, selected_crime=crime_dropdown);
```

**Insight**

The bar chart shows the Top 10 States by Rape cases, with Madhya Pradesh reporting the highest number of cases, significantly more than the other states. The top three states—Madhya Pradesh, Uttar Pradesh, and Rajasthan—lead by a large margin compared to the rest. The number of cases gradually decreases as we move down the list, with Bihar reporting the least number of cases among the top 10.

### ***Chart 3***

### **Visualization of States with the Lowest Crime Incidents**

```python
from ipywidgets import interact, Dropdown

# Define the crime categories for analysis
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
       'Assault against Women', 'Assault against Modesty of Women',
       'Domestic Violence', 'Women Trafficking']

# Calculate the total number of crimes for each state by summing the defined crime categories
crimes_df['Total Crimes'] = crimes_df[crime_columns].sum(axis=1)

# Group the data by 'State' and calculate total crimes per state, sorting in ascending order to find states with the lowest crime rates
state_crime_totals = crimes_df.groupby('State')['Total Crimes'].sum().sort_values(ascending=True)

# Get the bottom 10 states with the lowest total crimes
bottom_10_states = state_crime_totals.head(10).index

# Filter the data to include only the records from the bottom 10 states
bottom_10_data = crimes_df[crimes_df['State'].isin(bottom_10_states)]

# Function to update the bar plot based on the selected crime type from the dropdown
def update_barplot(selected_crime):
    # Group the data by 'State' and sum up the selected crime for the bottom 10 states
    bottom_10_crimes = bottom_10_data.groupby('State')[selected_crime].sum().sort_values(ascending=True)

    # Create a bar plot for the selected crime type in the bottom 10 states
    plt.figure(figsize=(12, 6))
    bottom_10_crimes.plot(kind='bar', color='green')
    plt.title(f'Bottom 10 States by {selected_crime}', fontsize=14)  # Set the title of the plot
    plt.xlabel('State', fontsize=12)  # Label for the x-axis
    plt.ylabel(f'Total {selected_crime} Cases', fontsize=12)  # Label for the y-axis
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.show()  # Display the plot

# Create a dropdown widget for selecting crime types, with 'Rape' set as the default value
crime_dropdown = Dropdown(
    options=crime_columns,  # List of crime categories as options for the dropdown
    value='Rape',           # Default crime type selected in the dropdown
    description='Crime Type:',  # Label for the dropdown
)

# Create an interactive plot that updates based on the selected crime type from the dropdown
interact(update_barplot, selected_crime=crime_dropdown);
```

**Insights:**

 The bar chart displays the bottom 10 states in India ranked by the total number of rape cases. Meghalaya has the highest number of cases among these states, nearing 2,000, while Lakshadweep reports the lowest, with almost no cases. The other states, including Daman and Diu, Puducherry, and Manipur, show a gradual increase in the number of reported rape cases, with Arunachal Pradesh and Mizoram nearing 1,000 cases. The data highlights significant variation in reported rape cases across these states.

### ***Chart 4***
### **Distribution of Crime Types in Top 10 States**

```python

# Define the crime categories to analyze
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                 'Assault against Women', 'Assault against Modesty of Women',
                 'Domestic Violence', 'Women Trafficking']

# Calculate total crimes for each state by summing the specified crime categories
crimes_df['Total Crimes'] = crimes_df[crime_columns].sum(axis=1)

# Group data by state and sum up the crimes for each crime type
state_crime_distribution = crimes_df.groupby('State')[crime_columns].sum()

# Sort the states by total crimes and select the top 10 states
top_10_states = crimes_df.groupby('State')['Total Crimes'].sum().sort_values(ascending=False).head(10).index
top_10_crime_distribution = state_crime_distribution.loc[top_10_states]

# Plot a stacked bar chart for the distribution of crime types in the top 10 states
top_10_crime_distribution.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title('Distribution of Crime Types in Top 10 States', fontsize=14)  # Title of the plot
plt.xlabel('State', fontsize=12)  # Label for the x-axis
plt.ylabel('Number of Crimes', fontsize=12)  # Label for the y-axis
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Legend placement
plt.show()  # Display the plot

```

### ***Chart 6***
### **Crime Distribution in Top 10 Years**

```python

# Calculate total crimes for each year
crimes_df['Total Crimes'] = crimes_df[crime_columns].sum(axis=1)
total_crimes_by_year = crimes_df.groupby('Year')['Total Crimes'].sum()

# Get the top 10 years with the highest total crimes
top_10_years = total_crimes_by_year.sort_values(ascending=False).head(10).index

# Filter the dataset for only the top 10 years
top_10_years_data = crimes_df[crimes_df['Year'].isin(top_10_years)]

# Melt the dataframe to long format for better visualization
crime_melted = top_10_years_data.melt(id_vars=['Year'], value_vars=crime_columns, var_name='Crime Type', value_name='Crime Count')

# Create a bar plot for crime distribution in the top 10 years
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Crime Count', hue='Crime Type', data=crime_melted, ci=None)  # Use ci=None to disable confidence intervals
plt.title('Crime Distribution in Top 10 Years', fontsize=14)  # Title of the plot
plt.xlabel('Year', fontsize=12)  # Label for the x-axis
plt.ylabel('Crime Count', fontsize=12)  # Label for the y-axis
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Legend placement
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()  # Display the plot
```

### **Insight**

The bar chart shows the distribution of various crimes against women over the past 10 years. Domestic violence (brown) consistently has the highest count each year, followed by kidnapping and abduction (orange) and assault against women (red). Rape (blue) remains relatively lower in comparison. The overall crime trend appears steady, with slight fluctuations year to year.

### ***Chart 7***

### **Proportion of Different Crimes Nationwide**

```python
# Proportion of Different Crimes Nationwide with Exploded Pie Chart

# Define the list of crime columns to analyze
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                 'Assault against Women', 'Assault against Modesty of Women',
                 'Domestic Violence', 'Women Trafficking']

# Calculate the total number of each crime type across all states and years
crime_totals = crimes_df[crime_columns].sum()

# Create an "explode" array to separate the largest crime segment for emphasis
explode = [0.1 if i == crime_totals.max() else 0 for i in crime_totals]

# Create a custom color palette with a more vibrant set of colors
custom_colors = sns.color_palette('Set3')

# Plot the pie chart with an exploded view for the largest crime category
plt.figure(figsize=(10, 8))  # Increase the figure size
plt.pie(crime_totals, labels=crime_totals.index, autopct='%1.1f%%', explode=explode, colors=custom_colors,
        startangle=140, shadow=True, wedgeprops={'edgecolor': 'black'})  # Add shadow and startangle for a unique look
plt.title('Nationwide Distribution of Crimes Against Women (Exploded View)', fontsize=14)
plt.show()  # Display the plot

```

**Insight**

  The pie chart shows the distribution of crimes against women in India. Domestic violence is the most prevalent crime, accounting for nearly 40% of cases. Other significant crimes include assault against women, assault against the modesty of women, and kidnapping and abduction.

```python
import matplotlib.pyplot as plt

# Filter the data for the year 2021
crimes_2021 = crimes_df[crimes_df['Year'] == 2021]

# Group by state and calculate total crimes in 2021
statewise_total_crimes_2021 = crimes_2021.groupby('State')['Total Crimes'].sum().sort_values(ascending=False)

# Get the top 10 and bottom 10 states by total crimes
top_10_states = statewise_total_crimes_2021.head(10)  # Top 10 states with the highest total crimes
bottom_10_states = statewise_total_crimes_2021.tail(10).iloc[::-1]  # Bottom 10 states with the lowest total crimes, reversed

# Create a figure for subplots to display both top 10 and bottom 10 states
plt.figure(figsize=(14, 6))  # Increase the figure size for better visibility

# Plot for top 10 states
plt.subplot(1, 2, 1)  # Define the position of the first subplot (1 row, 2 columns, 1st plot)
top_10_states.plot(kind='bar', color='blue')  # Create a bar plot for top 10 states
plt.title('Top 10 States by Total Crimes in 2021', fontsize=14)  # Set title for the plot
plt.xlabel('State', fontsize=12)  # Label for x-axis
plt.ylabel('Total Crimes', fontsize=12)  # Label for y-axis
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', linewidth=0.7)  # Add grid lines for y-axis

# Plot for bottom 10 states, with the order reversed
plt.subplot(1, 2, 2)  # Define the position of the second subplot (1 row, 2 columns, 2nd plot)
bottom_10_states.plot(kind='bar', color='red')  # Create a bar plot for bottom 10 states
plt.title('Bottom 10 States by Total Crimes in 2021 (Reversed Order)', fontsize=14)  # Set title for the plot
plt.xlabel('State', fontsize=12)  # Label for x-axis
plt.ylabel('Total Crimes', fontsize=12)  # Label for y-axis
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', linewidth=0.7)  # Add grid lines for y-axis

# Adjust layout to prevent overlap of subplots
plt.tight_layout(pad=3.0)  # Increase padding between subplots
plt.show()  # Display the plots
```

**Insight**

The bar charts illustrate the top and bottom 10 Indian states based on the total number of crimes reported in 2021. Tripura recorded the highest number of crimes, followed by Punjab and Madhya Pradesh, while Lakshadweep reported the lowest. Maharashtra, Meghalaya, and West Bengal also appear in the bottom 10, indicating a relatively lower crime rate compared to the top states.

### ***Chart 10***
### **Crime Types of different sates 2021**

```python
from ipywidgets import interact, Dropdown  # Importing necessary libraries for interactive widgets

# Assuming crimes_df is already loaded and contains the crime data

# Define the crime categories to analyze
crime_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths',
                 'Assault against Women', 'Assault against Modesty of Women',
                 'Domestic Violence', 'Women Trafficking']

# Function to update the bar plot based on the selected state
def update_barplot(selected_state):
    # Filter the data for the selected state
    state_data = crimes_df[crimes_df['State'] == selected_state][crime_columns].sum()  # Summing the total crimes for each category in the selected state

    # Plot the bar chart
    plt.figure(figsize=(10, 6))  # Set the figure size for the plot
    state_data.plot(kind='bar', color='Teal')  # Create a bar chart with red bars
    plt.title(f'Crime Types in {selected_state} for 2021', fontsize=14)  # Set the title of the plot
    plt.xlabel('Crime Type', fontsize=12)  # Label for x-axis
    plt.ylabel('Total Crimes', fontsize=12)  # Label for y-axis
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.grid(axis='y', linestyle='--', linewidth=0.7, color='gray')  # Optional grid for better readability
    plt.show()  # Display the plot

# Create a dropdown widget for selecting states
state_dropdown = Dropdown(
    options=crimes_df['State'].unique(),  # Get unique states from the dataframe for dropdown options
    value=crimes_df['State'].unique()[0],  # Set default value to the first state in the list
    description='Select State:',  # Label for the dropdown
)

# Create an interactive plot with the dropdown
interact(update_barplot, selected_state=state_dropdown);  # Link the dropdown selection to the update function
```

**Insight**

The bar chart illustrates the distribution of crimes against women in Andhra Pradesh in 2021. Domestic violence is the most prevalent crime, followed by assault against women and kidnapping and abduction. Rape and dowry deaths, while less frequent, are still concerning issues in the state. Women trafficking, though the least prevalent crime, remains a significant concern.
